```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import io
import statsmodels.api as sm
from pmdarima.arima import auto_arima

df_cpi = pd.read_stata(io.BytesIO(uploaded['cpi.dta']))

y = df_cpi['cpi']
model = auto_arima(y,
                  trace=True,
                  seasonal=False,
                  max_p=5,
                  max_q=5,
                  d=1,
                  start_p=0,
                  start_q=0,
                  start_P=0,
                  start_Q=0,
                  D=0,
                  m=1,
                  information_criterion='aic',
                  stepwise=True)
print(model.summary())
```
